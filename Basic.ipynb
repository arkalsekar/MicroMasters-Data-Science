```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import stats
```

```python
# Line Plot
x = np.arange(1, 100)
y = 2 * x
plt.plot(x, y)
plt.title("Line Graph")
plt.show()
```

```python
# bar graph 
x1 = [4, 6, 8, 10]
y1 = [1, 3, 15, 12]

x2 = [12,3, 73, 98] 
y2 = [12, 43, 98, 23]

plt.bar(x1, y1)
plt.bar(x2, y2, color='r')
plt.title("Bar Graph")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
x3 = [4, 6, 8, 10]
y3 = [1, 3, 15, 12]
plt.scatter(x3, y3)
plt.title("Scatter Graph")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```
